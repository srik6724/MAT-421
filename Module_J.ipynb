4.1 - Network Analysis



*  Network Analysis is important in that it is applicable to many problems involving large amounts of data.

*   Correlation networks are based on pairwise correlations between variables and are used in many commercial applications like data mining.

*  Social Graph: A graphical model of a network
    * Nodes: Individuals within a network
    * Edge: Connects two individuals related by the characterizing relationship
    * Can be/is used to model the userbase of social media platforms

Clustering; Community Detection:
*   Discovering groups of nodes more connected with one another within the group than without.
*   Used in practice to generate friend reccomendations and collaborative filtering.

Graph Models:



* Nodes: Objects modeled in a graph
  *  Degree: Number of edges connected to a node
* Edges: Connections between nodes
* Graph: G(V, E)
  * V = the set of nodes
  * E = the set of edges
  * Directed: edges have direction
* Path: Sequence of edges where nodes and edges are distinct
  * Closed Path: Cycle
  * Length: Number of edges traversed
* N-hop neighborhood: The set of nodes within n hops distance from a node v
* Python has the 'networkx' library that enables easy plotting of graphs:










```python
# An example of using the networkx library to plot a simple, directed graph

import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)

# Add edges
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(5, 3)
G.add_edge(1, 5)
G.add_edge(4, 5)

# Plot
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.title("Directed Graph")
plt.show()
```

4.2 Spectral Graph Bipartitioning

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create graph
G = nx.Graph()
# Add nodes
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)

# Add edges
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(5, 3)
G.add_edge(1, 5)
G.add_edge(4, 5)

# Define cut
cut_edges = [(1, 2), (1, 3), (2, 3)]

# remove cut edges
G_cut = G.copy()
G_cut.remove_edges_from(cut_edges)

# Plot the original graph
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.title("Original Graph")

# Plot the cut
plt.subplot(1, 2, 2)
nx.draw(G_cut, pos, with_labels=True)
plt.title("Graph after Cut")

plt.tight_layout()
plt.show()
```
