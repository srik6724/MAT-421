17.1 Interpolation Problem Statement

We are given a set of independent values xi and dependent values yi where i = 1...n. Similar to regression, the goal of interpolation is to find the estimation function for these given set of points. Interpolation does not require a model to be fit for the data, but it does talk about the quality of the interpolation.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use Seaborn to set "poster" style context
sns.set_context("poster")

x = [1, 2, 3, 4]
y = [1, 5, 3, 4]

plt.figure(figsize=(10, 8))
plt.plot(x, y, 'bo', label='Data Points')
plt.plot(2.5, 4, 'ro', label='Interpolated Point')
plt.title('Interpolation data point')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

17.2 Linear Interpolation

Assume that x data-points are in ascending order. Let x be a point such xi < x < xi+1. Then, the linear interpolation is calculated as y(x) = yi + ((yi-1-yi)(x-xi) / (xi+1 - xi))

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import seaborn as sns

# Use Seaborn to set "poster" style context
sns.set_context("poster")

x = [0, 1, 2]
y = [1, 3, 2]

f = interp1d(x, y)
y_hat = f(1.5)
print(y_hat)

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(1.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Cubic spine interpolation is a set of piecewise cubic functions between the boundaries of (xi, yi) and (xi+1, yi+1). The cubic function is Si(x) = aix^3 + bix^2 + cix + d, valid for xi <= x <= xi+1 for i = 1, ..., n-1. We need to find the four coefficients ai, bi, ci, and d to determine the interpolating function. For n points, we have 4(n-1) cubic functions to find.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Use Seaborn to set "poster" style context
sns.set_context("poster")
x = [0, 1, 2]
y = [1, 3, 2]
# use bc_type = 'natural' adds the constraints as we described above
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)
plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
